```python
import pandas as pd
import numpy as np
import zipfile as zp
zf = zp.ZipFile('external_datasets.zip')
#read in data and change into dataframe
cities_df = pd.read_csv(zf.open("external_datasets/cities_2010.csv"), encoding = "ISO-8859-1")
house_df = pd.read_csv(zf.open("external_datasets/housing.csv"), encoding = "ISO-8859-1")
income_df = pd.read_csv(zf.open("external_datasets/income.csv"), encoding = "ISO-8859-1")
poverty_df = pd.read_csv(zf.open("external_datasets/poverty.csv"), encoding = "ISO-8859-1")
age_df = pd.read_csv(zf.open("external_datasets/median_age.csv"), encoding = "ISO-8859-1")

# rename features
income_df.columns = ['GEO.id', 'GEO.id2', 'GEO.display-label', 'GCT_STUB.target-geo-id',
       'GCT_STUB.target-geo-id2', 'GCT_STUB.display-label',
       'GCT_STUB.display-label.1', 'income', 'income_margin_error']
poverty_df.columns = ['GEO.id', 'GEO.id2', 'GEO.display-label', 'GCT_STUB.target-geo-id',
       'GCT_STUB.target-geo-id2', 'GCT_STUB.display-label',
       'GCT_STUB.display-label.1', 'poverty_rate', 'poverty_margin_error']
age_df.columns = ['GEO.id', 'GEO.id2', 'GEO.display-label', 'GCT_STUB.target-geo-id',
       'GCT_STUB.target-geo-id2', 'GCT_STUB.display-label',
       'GCT_STUB.display-label.1', 'Median_age', 'age_margin_error']
cities_df.columns = ['GEO.id', 'GEO.id2', 'GEO.display-label', 'GCT_STUB.target-geo-id',
       'GCT_STUB.target-geo-id2', 'GCT_STUB.display-label',
       'GCT_STUB.display-label.1', 'Population', 'Housing units', 'Area in square miles - Total area', 'Area in square miles - Water area',
       'Area in square miles - Land area', 'Density per square mile of land area - Population', 'Density per square mile of land area - Housing units']

# merge dataframe
def merge_gis_data(df_list):
    final = df_list[0]
    for i in range(1,len(df_list)):
        final = pd.merge(final,df_list[i], how='outer', on='GCT_STUB.target-geo-id')
    return final

df_list = [cities_df,income_df,poverty_df,age_df]
finalized_df = merge_gis_data(df_list)
```
